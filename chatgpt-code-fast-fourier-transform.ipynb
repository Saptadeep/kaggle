```python
# Using Fast Fourier Transform (FFT) algorithm in Python to transform a time-domain signal into 
# its frequency domain representation:
# See detailed exxplanation notes at end
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample signal
t = np.linspace(0, 1, 1000)
f = 5  # frequency in Hz
signal = np.sin(2 * np.pi * f * t) + 0.5 * np.sin(6 * 2 * np.pi * f * t)

# Perform FFT
fft = np.fft.fft(signal)

# Calculate the corresponding frequencies
freqs = np.fft.fftfreq(len(signal), t[1] - t[0])

# Plot the results
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].plot(t, signal)
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Amplitude')
axs[1].plot(freqs, np.abs(fft))
axs[1].set_xlabel('Frequency [Hz]')
axs[1].set_ylabel('Magnitude')
plt.show()

# Notes
# First we  generate a sample signal that is a combination of two sine waves with different frequencies.
# Then, we use the fft function from the numpy.fft module to compute the FFT of the signal.
# Next, we use the fftfreq function to calculate the corresponding frequencies for each point in the FFT output. 
# Finally, we plot the original signal in the time domain and the FFT result in the frequency domain.
# The output should be a plot with two subplots.
# The first subplot shows the original signal in the time domain, while the second subplot shows the FFT result
# in the frequency domain. The frequency peaks in the FFT plot should correspond to the frequencies of the 
# two sine waves in the original signal.
```
